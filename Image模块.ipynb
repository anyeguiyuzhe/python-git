### 2.Mode类

| modes | 描述 |
| :-: | :-: |
| L | 8位像素，黑白 |
| RGB | 3x8位像素，真彩 |
| RGBA | 4x8位像素，真彩+透明通道 |


```python
photo_a=photo.convert('RGBA')
photo_a.show()
#保存图片 photo_a.save(path)
```

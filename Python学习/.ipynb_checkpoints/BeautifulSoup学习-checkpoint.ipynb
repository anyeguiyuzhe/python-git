# 1. 节点选择器

直接调用节点的名称就可以选择节点元素，再调用string属性就可以得到节点内的文本

```python
soup.title
soup.head
soup.head.string
#嵌套选择
#比如我们获取了head节点元素，可以继续调用head来获取其内部的title元素
soup.head.title
```

### 字节点和子孙节点


```python
#选取节点元素之后，如果想要获取它的直接子节点，可以调用contents属性
soup.head
soup.head.contents#返回列表,其中的每个元素都是p节点的直接子节点

#子孙节点,soup.head.children返回的是iterator类型
for x,y in enumerate(soup.head.chileren):
    print((x,y))

#父节点和祖先节点
#调用parent属性
html = """
<html>
    <head>
        <title>The Dormouse's story</title>
    </head>
    <body>
        <p class="story">
            Once upon a time there were three little sisters; and their names were
            <a href="http://example.com/elsie" class="sister" id="link1">
                <span>Elsie</span>
            </a>
        </p>
        <p class="story">...</p>
"""
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
print(soup.a.parent)
#parent仅仅返回a节点的直接父节点而没有在外寻找父节点的祖先节点
#调用parents获取所有祖先节点
for x,y in enumerate(list(soup.a.parents)):
    print(x,y)
```

# 2 方法选择器

### 1.根据节点名称来查询元素

```python
#find_all查询所有复合条件的元素
#API:find_all(name,attrs,recursive,text,**kwargs)
html='''
<div class="panel">
    <div class="panel-heading">
        <h4>Hello</h4>
    </div>
    <div class="panel-body">
        <ul class="list" id="list-1">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
            <li class="element">Jay</li>
        </ul>
        <ul class="list list-small" id="list-2">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
        </ul>
    </div>
</div>
'''
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
#查询的结果依然是bs4.element.Tag类型
print(soup.find_all(name='ul'))
print(type(soup.find_all(name='ul')[0]))
#因为都是Tag类型，所以依然可以进行嵌套查询
for ul in soup.find_all('ul'):
    print(ul.find_all('li'))

#遍历每个li，获取文本
for ul in soup.find_all(name='ul'):
    print(ul.find_all('li'))
    for li in ul.find_all(name='li'):
        print(li.string)
```

### 2.传入一些属性来查询

```python
html='''
<div class="panel">
    <div class="panel-heading">
        <h4>Hello</h4>
    </div>
    <div class="panel-body">
        <ul class="list" id="list-1" name="elements">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
            <li class="element">Jay</li>
        </ul>
        <ul class="list list-small" id="list-2">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
        </ul>
    </div>
</div>
'''
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
print(soup.find_all(attrs={'id': 'list-1'}))
print(soup.find_all(attrs={'name': 'elements'}))
#对于一些常见的属性，如id和class，可以不用attrs来传递
print(soup.find_all(id='list-1'))
#由于class是python的关键字，所以需要加一个下划线
print(soup.find_all(class_='element'))
```

### 3. text参数用来匹配节点的文本,传入的形式可以是字符串，可以是正则表达式对象

```python
import re
html='''
<div class="panel">
    <div class="panel-body">
        <a>Hello, this is a link</a>
        <a>Hello, this is a link, too</a>
    </div>
</div>
'''
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
print(soup.find_all(text=re.compile('link')))
```

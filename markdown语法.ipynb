# <center>MarkDown语法</center>

1. 如果你想引用一段文字，那就使用'>'

 > 这是引用的文字

2. 强调语法，在文字前后加两个'*'号

 *这是需要强调的话*

3. 粗体,加两个*

 **文字加粗**

4. 斜体，在文字后加一个*号

 *这是斜体文字*

5. 下划线，在文字前后加一个_

 _文字下面划线了_

6. 删除线，在文字前后加一个波浪线

 ~~这是错误的语句~~·

7. 插入链接，如果你想把一段文字加上链接指向到某个网址，就把文字用中括号围起来'[]'，然后再加上一个小括号'()'

 [这是python官网的链接](https://www.python.org)

8. 插入图片，和插入链接很像，就是在前面加了一个感叹号'!'

 ![百度搜索](http://www.baidu.com/img/bd_logo1.png)

9. 有序列表：数字后加点，再空一格写内容

10. 无序列表，只需要在文字前面加上'-'就行

 - 下载python
 - 安装anaconda

11. 换行，段落与段落之间需要强制空一格才能分段

12. 分隔线，你可以在一行中使用三个'*'来创建分隔线

 分隔内容

## 1.1 股票列表

```python
import tushare as ts
pro=ts.pro_api()
import pandas as pd
import numpy as np
```

- 输入参数
|名称|描述|
|:-:|:-:|
|is_hs|是否深港通标的,N否，H沪股通，S深股通|
|list_status|上市状态：L上市，D退市，P暂停上市|
|exchange|交易所：SSE上交所，SZSE深交所，HKEX港交所|

- 输出参数
|名称|描述|
|:-:|:-:|
|ts_code|TS代码|
|symbol|股票代码|
|name|股票名称|
|area|所在地域|
|industry|行业|
|fullname|全称|
|enname|英文全称|
|market|市场类型（主板/中小板/创业板/科创板)|
|exchange|交易所代码|

```python
#查询当前正常上市交易的股票列表
x=pro.stock_basic(list_status='l',fields='ts_code,symbol,name,area,industry,list_date')
x.head()
```
